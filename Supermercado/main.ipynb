```python
import pandas as pd
from faker import Faker
import random
import datetime
import google.generativeai as genai



fake = Faker()

# Definindo a data atual
today = datetime.date.today()
# Definindo o número de anos para a frente que queremos gerar uma data de validade
years_ahead = 5

# Definindo a data final como a data atual mais alguns anos à frente
end_date = today + datetime.timedelta(days=365 * years_ahead)

# Lista de produtos de supermercado
produtos_supermercado = ['Arroz', 'Feijão', 'Macarrão', 'Leite', 'Café', 'Azeite', 'Sal', 'Açúcar', 'Farinha', 'Óleo', 'Carne', 'Frango', 'Peixe', 'Ovos', 'Pão', 'Queijo', 'Presunto', 'Manteiga', 'Iogurte', 'Refrigerante']

# Função para gerar dados fictícios 
def gerar_dados_ficticios(n):
    dados = []
    for i in range(n):
        item = f'Item {i+1}'
        descricao = ', '.join(random.choices(produtos_supermercado))
        data_validade = today + datetime.timedelta(days=random.randint(0, (end_date - today).days))
        preco = round(random.uniform(1.0, 20.0), 2)
        entrada = random.randint(10, 100)
        saida = random.randint(0, entrada)
        estoque = entrada - saida
        filial = random.randint(1, 5)
        dados.append([item, descricao,data_validade, entrada, saida, preco, estoque, filial])
    return dados

# Gerar 20 linhas de dados fictícios
dados = gerar_dados_ficticios(100)

# Criar DataFrame com os dados
df = pd.DataFrame(dados, columns=['Item', 'Descrição','data_validade', 'Entrada', 'Saída', 'Preço', 'Estoque', 'Filial'])

# Salvar DataFrame como uma planilha Excel
df.to_excel('dados_ficticios.xlsx', index=False)
```

```python
import openpyxl

# Função para ler dados de um arquivo Excel

def ler_dados_excel(df):
    workbook = openpyxl.load_workbook(df)
    sheet = workbook.active
    df = []
    for row in sheet.iter_rows(values_only=True):
        df.append({
            "nome": row[0],
            "preco": row[1],
            "quantidade": row[2],
            "data_validade": row[3]
        })
    return df
```
